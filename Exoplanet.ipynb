```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a dataframe planet1
p1 = {'Spectral Type': ['G'], 'Age of the System (Gyr)': [8.16], 'Binary System': ['Single Star'], 'Stellar Luminosities (Solar Units)': [1.32], 'Total Stellar Mass (Solar Masses)': [1.07], 'Stellar Radius (Solar Radii)': [1.07], 'Planet Orbital Period (Years)': [1.73], 'Planet Semi-Major Axis (AU)': [1.47], 'Planet Mass (Earth Masses)':[3.16], 'Planet Radius (Earth Radii)': [1.41], 'Eccentricity': [0.0954], 'N2': [84.376], 'CO2':[2.721], 'H2O': [0.511], 'O2': [12.393], 'Others':[0.000], 
      'Number of other (known) Planets in the System': [3], 'Location in the Galaxy': ['Disk'], 'Distance from Earth (lightyears)': [63.9], 'Atmospheric Pressure (atm)': [1.65], 'Planetary Albedo':[0.374], 'Planet Axial Tilt (degrees)': [21.2], 'Planet Rotation Period (Hours)': [20.2], 'Magnetic Field Strength (Units of Earth Magnetic Field)': [0.000663]}

planet1 = pd.DataFrame(data=p1)

planet1
```

```python
# Create a dataframe planet2
p2 = {'Spectral Type': ['F'], 'Age of the System (Gyr)': [0.0568], 'Binary System': ['Single Star'], 'Stellar Luminosities (Solar Units)': [4.49], 'Total Stellar Mass (Solar Masses)': [1.46], 'Stellar Radius (Solar Radii)': [1.43], 'Planet Orbital Period (Years)': [4.81], 'Planet Semi-Major Axis (AU)': [3.23], 'Planet Mass (Earth Masses)':[3.07], 'Planet Radius (Earth Radii)': [1.47], 'Eccentricity': [0.11], 'N2': [88.003], 'CO2':[1.897], 'H2O': [4.838], 'O2': [5.261], 'Others':[0.000], 
      'Number of other (known) Planets in the System': [3], 'Location in the Galaxy': ['Open Cluster'], 'Distance from Earth (lightyears)': [1.58E+03], 'Atmospheric Pressure (atm)': [0.93], 'Magnetic Field Strength (Units of Earth Magnetic Field)': [1.2]}

planet2 = pd.DataFrame(data=p2)

planet2
```

```python
# Add a 'Planet' column to identify which planet each row belongs to
planet1.insert(0, "Planet", "Planet 1")
planet2.insert(0, "Planet", "Planet 2")

# Combine the two dataframes
planets = pd.concat([planet1, planet2], ignore_index=True)
```

```python
# Add a column indicating whether each planet is a terrestrial planet
planets['Terrestrial Planet'] = np.where((planets['Planet Radius (Earth Radii)'] >= 0.5) & (planets['Planet Radius (Earth Radii)'] <= 2), 'Yes', 'No')

planets
```

```python
# Add the habitable zone columns to the DataFrame
def habitable_zone(L):
    inner_bound = np.sqrt(L/1.1)
    outer_bound = np.sqrt(L/0.53)
    return inner_bound, outer_bound

planets['Inner Bound of Habitable Zone (AU)'], planets['Outer Bound of Habitable Zone (AU)'] = habitable_zone(planets['Stellar Luminosities (Solar Units)'])

planets
```

```python
# Add a column indicating whether each planet is in the habitable zone
planets['In Habitable Zone'] = np.where((planets['Planet Semi-Major Axis (AU)'] >= planets['Inner Bound of Habitable Zone (AU)']) & (planets['Planet Semi-Major Axis (AU)'] <= planets['Outer Bound of Habitable Zone (AU)']), 'Yes', 'No')

planets
```

```python
from pandas.core.groupby.generic import SingleManager
import scipy.constants as const

sigma = const.sigma  # Stefan-Boltzmann constant

# Define a function to calculate the equilibrium temperature
def equilibrium_temp(L, a, albedo):
    return np.sqrt(np.sqrt(L * (1 - albedo) / (16 * sigma)) * (1 / np.sqrt(a)))

# Add the 'Equilibrium Temperature (K)' column to the DataFrame
planets['Equilibrium Temperature (K)'] = equilibrium_temp(planets['Stellar Luminosities (Solar Units)'], planets['Planet Semi-Major Axis (AU)'], planets['Planetary Albedo'])

planets
```
